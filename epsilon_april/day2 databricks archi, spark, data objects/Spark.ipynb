```python
df=spark.read.json("/Volumes/dev/naval/raw/drivers.json")
```

```python
df1=(df
     .withColumn("ingestion_date",current_timestamp())
     .withColumn("path",col("_metadata.file_path"))
     .withColumn("forename",col("name.forename"))
     .withColumn("surname",col("name.surname"))
     .drop("name")
     )
```

```python
df1.write.mode("overwrite").saveAsTable("dev.naval.drivers")
```
